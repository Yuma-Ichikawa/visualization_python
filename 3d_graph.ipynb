# 3次元グラフを作成

matplotlibの『mpl_toolkits.mplot3d』を使用し3次元グラフを作成します。

3次元で描画する場合、２次元のメッシュを作成する必要があります。

```
# 使用するパッケージをインポート
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
```

```
# 図示する関数
t = np.linspace(- 2*np.pi, 2*np.pi)
x = np.cos(t)
y = np.sin(t)
z = t
```

```
fig, ax = plt.subplots(subplot_kw={'projection' : '3d'})
ax.set_xticks(np.linspace(-1.0, 1.0, 5))
ax.set_yticks(np.linspace(-1.0, 1.0, 5))
ax.plot(x, y, z, c='b')
```

### 曲面をプロット

* rstride, cstride : 各軸に対して何ステップごとにプロットするか
* cmap : カラーマップの指定
* alpha : 透明度を設定

```
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw={'projection' : '3d'})
ax.plot_surface(X, Y, Z, rstride=1, cstride=10, cmap='cividis', alpha=0.3)
plt.show()
```

#### カラーバーを設定する

* ax : カラーバーを設置するグラフ
* shrink : カラーバーのサイズ

```
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw={'projection' : '3d'})
surface = ax.plot_surface(X, Y, Z, rstride=1, cstride=10, cmap='cividis', alpha=0.3)
fig.colorbar(surface, ax=ax, shrink=0.5)
plt.show()
```

### ワイヤーフレーム

* linewidth : 線の太さ

```
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw={'projection' : '3d'})
ax.plot_wireframe(X, Y, Z, linewidth=0.3)
plt.show()
```

### scatter3D

* s : マーカの大きさ
* c : マーカの色
* depthshade : マーカに影をつける



```
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw={'projection' : '3d'})
ax.scatter3D(X, Y, Z, s=1, c='red', depthshade=True)
plt.show()
```

### 3次元グラフを回転

* elev : 縦方向の回転
* azim : 横方向の回転

```
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw={'projection' : '3d'})
ax.plot_surface(X, Y, Z, rstride=1, cstride=10, cmap='cividis', alpha=0.3)
ax.view_init(elev=10, azim=50)
plt.show()
```
