## ヒストグラムについて

sns.distplot利用することでヒストグラムが作成できる。

デフォルト : カーネル密度推定と呼ばれる方法で確率密度関数と正規化された（面積が１になるように変換された）ヒストグラム が表示される

Parameter : 

- kde : 確率密度関数の表示設定
- norm_hist : 正規化の設定

```
sns.distplot(a = titanic['age'])
plt.show()
```

```
sns.distplot(a = titanic['age'], kde=False, norm_hist=False)
plt.show()
```

### 散布図について

散布図は、joinplotとpirplotによって作成可能

joinplotに関して

Parameteer : 
- x : 横軸のカラム
- y : 縦軸のカラム
- data : 使用するデータフレーム
- alpha : 透明度
- kind : reg(回帰）

pirplotに関して



```
# データの範囲を制限
titanic = titanic[titanic['fare'] < 300]
```

```
sns.jointplot(x='age', y='fare', data=titanic, alpha=0.3)
plt.show()
```

```
# regressionを使用するときの透明度の設定方法
sns.jointplot(x='age', y='fare', data=titanic, kind ='reg', joint_kws={'scatter_kws': {'alpha':0.3}})
plt.show()
```

```
sns.jointplot(x='age', y='fare', data=titanic, kind='kde')
plt.show()
```

```
# pirplot
sns.pairplot(data = titanic[['age', 'fare']])
plt.show()
```

```
# 色合いをつける
sns.pairplot(titanic[['age', 'fare','survived']], 
             hue='survived', plot_kws={'alpha':0.3},
             diag_kind='hist')
```
